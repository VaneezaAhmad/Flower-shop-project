```python
import pandas as pd 
import csv
import numpy as np
import json
import matplotlib.pyplot as plt
import os
```

#  DataFrames for Silver Spring and Leesburg Flower Shops

```python
# Silver Spring
ss_flowers = pd.read_csv("silverspring-floral.csv")  #, sep='delimiter', header=None)
ss_flowers.replace(" ",np.nan, inplace=True)
ss_flowers.columns=["Business id", "Name", "Alias", "is_closed", "Longitude", "Latitude", "Price", "Review_count", "Rating"]
ss_flowers
```

```python
# Leesburg
lb_flowers = pd.read_csv("leesburg-floral.csv")
ss_flowers.replace(" ",np.nan, inplace=True)
lb_flowers.columns=["Business id", "Name", "Alias", "is_closed", "Longitude", "Latitude", "Price", "Review_count", "Rating"]
lb_flowers.head()
```

# Percentage of each price

```python
# Percentage of dollar signs in Leesburg
three = (100 * (lb_flowers['Price'] == "$$$").mean())
two = (100 * (lb_flowers['Price'] == "$$").mean())
one = (100 * (lb_flowers['Price'] == "$").mean())
zero = lb_flowers["Price"].isnull().mean() * 100

print("Percentage of three dollar signs :" , three)
print("Percentage of two dollar signs :" , two)
print("Percentage of one dollar signs :" , one)
print("Percentage of no signs :" , zero)
```

## Percentage of Each Yelp Rating by City

```python
ss_flowers.Rating.value_counts(normalize = True).mul(100).reindex([1, 1.5, 2, 3, 3.5, 4, 4.5, 5]).plot(kind="bar", figsize = (10,5), color = "blue")
plt.title("Distribution of Yelp Ratings in Silver Spring, MD")
plt.xlabel("Yelp Ratings")
plt.ylabel("Percentage")
```

```python
lb_flowers.Rating.value_counts(normalize = True).mul(100).reindex([1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]).plot(kind="bar", figsize = (10,5), color = "crimson")
plt.title("Distribution of Yelp Ratings in Leesburg, VA")
plt.xlabel("Yelp Ratings")
plt.ylabel("Percentage")
```
